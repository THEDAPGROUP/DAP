```python
# accessing the stocks data from mongoDB

from pymongo import MongoClient
client = MongoClient("mongodb://%s:%s@127.0.0.1" % ("dap", "dap"))
db = client['ca_database']
stocks = db.stocks
stocks_list = list(stocks.find())
stocks_list
```

```python
# Mysql connection 
import mysql.connector as connector

def connection():

    config = {
        "user": "root",
        "password": "root",
        "host": "localhost",
        "port": 3306,
        "database": "dap_ca"
    }
    try:
        c = connector.connect(**config)
        return c
    except (Exception) as dbError :
        print ("Error while connecting to Mysql", dbError)
        
def createDatabase(): #no error method
    cn = connection()
    cur = cn.cursor()
    cur.execute("create database IF NOT EXISTS dap_ca")

createDatabase()
```

```python
def createTable():
    cn = connection()
    cn.autocommit = True
    cur = cn.cursor()
    cur.execute('''CREATE TABLE IF NOT EXISTS Stocks (
                        stock_id int NOT NULL AUTO_INCREMENT PRIMARY KEY,
                        symbol varchar(10),
                        price FLOAT (15,5),
                        pchange FLOAT (15,5),
                        prev_close FLOAT (15,5),
                        open FLOAT (15,5),
                        day_range_open FLOAT (15,5),
                        day_range_close FLOAT (15,5),
                        volume FLOAT (15,5),
                        market_cap varchar(10),
                        date TIMESTAMP(0)
                    );''')
    cur.close()

createTable()
```

```python
def insert_stocks():
    cn = connection()
#     cn.autocommit(True)
    cur = cn.cursor()
    insertString = '''INSERT IGNORE INTO Stocks
(symbol, price, pchange, prev_close, open, day_range_open, day_range_close, volume, market_cap, date)
VALUES ('{}',{},{},{},{},{},{},{},'{}','{}');'''
    for stock in stocks_list:
        insert_stock = insertString.format(stock['symbol'], stock['price'], stock['change'], stock['prev_close'], 
                            stock['open'], stock['day_range'].split(" - ")[0], stock['day_range'].split(" - ")[1], 
                            stock['volume'].replace(',', ''), stock['market_cap'], stock['date'].strftime("%Y-%m-%d %H:%M:%S"))
        print(insert_stock)
        try:
            cur.execute(insert_stock)
        except:
            print("continuing with other records")
    cn.commit()
    cur.close()

insert_stocks()
    
```
